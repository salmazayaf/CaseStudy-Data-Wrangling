#NOMOR 1

```python
# Code untuk Soal No 01 diberikan sebagai Contoh dan Bonus
# Pelajari dengan baik

nilai_dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1) 
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN',  right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]
sb = nilai.groupby(nilai['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_dosen = nilai_dosen.set_index(['NIDN']) 
nilai_dosen = pd.concat([nilai_dosen, sb], axis=1)
nilai_dosen
```

## Interpretasi 01: 

Dapat dilihat dari tabel diatas bahwa dosen yang rata-rata nilai mahasiswanya paling tinggi adalah Andri Basuki dengan rata-rata 85,75. Dilihat dari standar deviasi juga memiliki nilai yang paling kecil diantara dosen-dosen lain sehingga mahasiswa-mahasiswa yang diampunya memiliki jarak nilai yang tidak jauh satu sama lain. Nilai mean cukup merepresentasikan data nilainya.

```python
# Code Untuk Soal No 02

nilai_dosen2 = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen2 = nilai_dosen2.rename({'nilai': 'mean_nilai'}, axis=1) 
nilai_dosen2 = pd.merge(dosen, nilai_dosen2, left_on='NIDN', right_index=True)
nilai_dosen2 = nilai_dosen2[['NIDN', 'nama','Pendidikan_Terakhir','mean_nilai']]
sb = nilai.groupby(nilai['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_dosen2 = nilai_dosen2.set_index(['NIDN']) 
nilai_dosen2 = pd.concat([nilai_dosen2, sb], axis=1) 
nilai_dosen2 = nilai_dosen2.set_index('Pendidikan_Terakhir')
nilai_dosen2
```

## Interpretasi 02:

Berdasarkan tabel diatas, jika mengurutkan rata-rata nilai yang diberikan dosen dengan pendidikan terakhirnya, Dosen bernama Andi Basuki dan zulfikar dari empat dosen merupakan dua dosen yang memiliki rata-rata nilai yang lebih tinggi dibanding kedua dosen lainnya. Andi basuki dan zulfikar juga berpendidikan terakhir S3 sedangkan kedua dosen lainnya berpendidikan terakhir S2. Jadi, dapat disimpulkan bahwa dari data yang tersedia, dosen dengan pendidikan terakhir S3 menghasilkan mahasiswa dengan nilai yang lebih baik.

```python
nilai_angkatan = nilai.groupby(nilai['nim']).mean()[['nilai']]
nilai_angkatan = nilai_angkatan.rename({'nilai': 'mean_nilai'}, axis=1)
nilai_angkatan = pd.merge(mahasiswa, nilai_angkatan, left_on = 'nim', right_index = True)
nilai_angkatan = nilai_angkatan[['Angkatan', 'mean_nilai', 'nim']]
sb = nilai.groupby(nilai['nim']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_angkatan = nilai_angkatan.set_index(['nim']) 
nilai_angkatan = pd.concat([nilai_angkatan, sb], axis=1) 
nilai_angkatan = nilai_angkatan.pivot_table(index='Angkatan')

nilai_angkatan
```

## Interpretasi 05

Berdasarkan hasil pengolahan data, terlihat bahwa ada perbedaan prestasi akademik antar angkatan mahasiswa. Pada tabel hasil di atas, mahasiswa angkatan 2019 memiliki perolehan nilai rata-rata sebesar 75.59375 dengan standar deviasi 13.385601. Sementara itu, mahasiswa angkatan 2020 memiliki perolehan nilai rata-rata sebesar 73.00000 dengan standar deviasi 13.547877 yang artinya jarak nilai antar mahasiswa 2020 lebih jauh daripada angkatan 2019.
Dengan perolehan nilai rata-rata lebih tinggi dan standar deviasi yang lebih kecil, dapat disimpulkan bahwa prestasi akademik mahasiswa angkatan 2019 lebih baik dibandingkan mahasiswa angkatan 2020.
