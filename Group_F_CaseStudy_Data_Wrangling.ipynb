## Interpretasi 03:

Pertanyaan: Apakah terdapat perbedaan performa akademik antara mahasiswa yang berasal dari Madrasah/SMK dan SMU?

Yang kami lakukan pada kode adalah menyatukan tabel nilai dan mahasiswa menggunakan nim sebagai patokan dan kemudian membuang kolum yang menurut kami tidak diperlukan. Setelah itu nim dijadikan index agar ia tidak dianggap sebagai variabel yang memiliki nilai. Kemudian kami ingin mendeskripsikan data pada tabel berdasarkan mata kuliah dan asal sekolah. Menurut kami melakukan grouping berdasarkan mata kuliah perlu karena tiap mata kuliah memiliki konten yang berbeda sehingga bisa jadi asal sekolah berpengaruh pada nilai di mata kuliah tertentu. 

Dibanding hanya melakukan pencarian mean dan standar deviasi kami kira penting untuk mencari statistik deskriptif dari data. Mengapa? Karena hanya memiliki kedua data tersebut bisa jadi tidak memberikan kami gambaran lengkap mengenai performa akademik siswa berdasarkan asal sekolah pada mata kuliah tertentu. Misal, pada mata kuliah CS01, pada anak-anak yang berasal dari SMK, ada satu anak yang mengangkat rata-rata nilai dari grup tersebut dan perbedaan antar grup juga tidak terekfleksikan oleh standar deviasi yang cenderung memberikan berat lebih kepada nilai yang paling besar. Walaupun, hal ini akan lebih relevan jika kami memiliki dataset yang lebih besar.

Perbedaan dalam grup masing-masing dapat kami lihat dari perbedaan min dan max, jarak antar kuartil, dan juga standar deviasi. Hal ini penting guna membandingkan persebaran kemampuan yang terjadi antara beda-beda grup. Kami dapat melihat asal sekolah mana yang menghasilkan mahasiswa-mahasiswi dengan kemampuan yang merata dan yang memiliki kemampuan yang jomplang . Salah satu cara mengukur performa akademik antara mereka yang berasal dari sekolah berbeda adalah dengan membandingkan nilai mean grup satu dengan lainnya. 

Untuk mata kuliah CS01, dapat kami lihat bahwa mereka yang berasal dari Madrasah memiliki rata-rata yang paling tinggi yaitu 90. Hanya terdapat dua orang yang berasal dari Madrasah yang mengambil mata kuliah CS01 dan keduanya memiliki nilai yang sama. Mereka yang berasal dari SMK dan SMU masing-masing ada tiga dan dua orang yang mengikuti CS01. Dapat dilihat bahwa terdapat perbedaan yang signifikan antara min dan max pada nilai-nilai mahasiswa yang berasal dari SMK dan SMU serta standar deviasi yang cukup tinggi. Sehingga dapat kami katakan perbedaan kemampuan dalam grup tersebut cukup signfikan. Berdasarkan data pada mata kuliah ini, mereka yang dari Madrasah memiliki kemampuan merata dan lebih unggul dari mereka dari SMU dan SMK.

Untuk mata kuliah KAL01, SMK memiliki nilai rata-rata paling tinggi dengan angka yang berbeda cukup jauh dari SMA dan Madrasah (masing-masing sekitar 18 dan sekitar 13) diikuti oleh Madrasah di posisi kedua serta SMU diakhir. Dengan meninjau tiga ukuran untuk melihat perbedaan dalam grup, kami katakan bahwa kemampuan pada mereka yang berasal dari SMA dan Madrasah memiliki persebaran yang cukup jauh untuk mata kuliah ini. Sementara untuk SMK persebaran tersebut kecil. Sehingga untuk KAL01, kemampuan mereka yang dari SMK lebih merata dan lebih unggul.

Pada mata kuliah KAL02, mereka yang berasal dari MA dan SMK hanya satu masing-masing. Dari SMU terdapat dua. Hasil deskripisi menunjukkan bahwa yang berasal dari SMK memiliki nilai yang paling tinggi. Sementara kedua orang yang berasal dari SMU, berdasarkan min dan max, memiliki kemampuan yang kurang lebih sama.

Pada mata kuliah STA02, mereka yang berasal dari SMK dan MA masing-masing dua, dengan menggunakan ukuran perbedaan mereka yang berasal dari SMK dan MA memiliki perbedaan performa signfikan diantara mereka.
```python
nilai_angkatan = nilai.groupby(nilai['nim']).mean()[['nilai']]
nilai_angkatan = nilai_angkatan.rename({'nilai': 'mean_nilai'}, axis=1)
nilai_angkatan = pd.merge(mahasiswa, nilai_angkatan, left_on = 'nim', right_index = True)
nilai_angkatan = nilai_angkatan[['Angkatan', 'mean_nilai', 'nim']]
sb = nilai.groupby(nilai['nim']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_angkatan = nilai_angkatan.set_index(['nim']) 
nilai_angkatan = pd.concat([nilai_angkatan, sb], axis=1) 
nilai_angkatan = nilai_angkatan.pivot_table(index='Angkatan')

nilai_angkatan
```

## Interpretasi 05

Berdasarkan hasil pengolahan data, terlihat bahwa ada perbedaan prestasi akademik antar angkatan mahasiswa. Pada tabel hasil di atas, mahasiswa angkatan 2019 memiliki perolehan nilai rata-rata sebesar 75.59375 dengan standar deviasi 13.385601. Sementara itu, mahasiswa angkatan 2020 memiliki perolehan nilai rata-rata sebesar 73.00000 dengan standar deviasi 13.547877 yang artinya jarak nilai antar mahasiswa 2020 lebih jauh daripada angkatan 2019. Dengan perolehan nilai rata-rata lebih tinggi dan standar deviasi yang lebih kecil, dapat disimpulkan bahwa prestasi akademik mahasiswa angkatan 2019 lebih baik dibandingkan mahasiswa angkatan 2020.
